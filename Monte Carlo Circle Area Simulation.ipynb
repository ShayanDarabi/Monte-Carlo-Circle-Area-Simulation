### Monte Carlo Circle Area Simulation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = pd.Series(np.random.rand(10000))
```

```python
y = pd.Series(np.random.rand(10000))
```

```python
df = pd.DataFrame(
        {"X":x,
         "Y":y,
        "In/Out": np.where(np.sqrt((x - 0.5)**2 + (y - 0.5)**2) <= 0.5, 1, 0)}
)
```

```python
df
```

```python
In = df[df["In/Out"] == 1]
Out = df[df["In/Out"] == 0]
```

```python
plt.figure(figsize=(15, 15))
plt.scatter(In["X"], In["Y"])
plt.scatter(Out["X"], Out["Y"])
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Monte Carlo Circle Area Simulation")
```

```python
circle_area_with_simulation = (df["In/Out"].sum()) / 10000
circle_area_with_simulation
```

```python
circle_area_without_simulation = np.pi * np.power(0.5, 2)
circle_area_without_simulation
```

As observed, with 10,000 repetitions of this process, the area obtained through Monte Carlo simulation (0.7863) is approximately equal to the area of the circle calculated using the circle formula (0.7853981633974483), with a good approximation.
